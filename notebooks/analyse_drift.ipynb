```python
import pandas as pd
import numpy as np
import evidently
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset
```

```python


print(f"ðŸš€ Lancement avec Evidently {evidently.__version__} (Mode Compatible)")

# 1. Chargement
try:
    reference = pd.read_csv('../data/application_train.csv')
    current = pd.read_csv('../data/application_test.csv')
except:
    # Plan B si les dossiers sont organisÃ©s diffÃ©remment
    reference = pd.read_csv('data/application_train.csv')
    current = pd.read_csv('data/application_test.csv')

# 2. Nettoyage (On ne garde que les chiffres pour Ã©viter tout bug)
reference = reference.drop(columns=['TARGET'], errors='ignore')
common_cols = [c for c in reference.columns if c in current.columns and c != 'SK_ID_CURR']

# On prend 3000 lignes de chiffres
ref_sample = reference[common_cols].sample(3000, random_state=42).select_dtypes(include=[np.number])
cur_sample = current[common_cols].sample(3000, random_state=42).select_dtypes(include=[np.number])

print(f"âœ… DonnÃ©es chargÃ©es.")

# 3. GÃ©nÃ©ration du Rapport
print("ðŸ“Š Calcul du Drift en cours...")
report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=ref_sample, current_data=cur_sample)

# 4. Sauvegarde
filename = "data_drift_report.html"
report.save_html(filename)

print(f"\nðŸŽ‰ VICTOIRE ! Le fichier '{filename}' est crÃ©Ã© dans ton dossier.")
```
